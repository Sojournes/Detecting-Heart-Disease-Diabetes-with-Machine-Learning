```python
import pandas as pd

data = {
    "Cholestrol":[190, 250, 220, 260, 210, 180, 240, 230, 270, 200],
    "BloodPressure":[120, 150, 130, 160, 125, 110, 145, 135, 170, 115]
}

df = pd.DataFrame(data)
```

```python
def check(Cholestrol,Blood_pressure):
  if Cholestrol > 240 or Blood_pressure > 140:
    return "risky"
  else:
    return "Normal"
```

```python
df["HealthCondition"] = df.apply(lambda row: check(row['Cholestrol'],row['BloodPressure']),axis=1)
```
